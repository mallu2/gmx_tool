### Analysis of GROMACS minimization and equilibration

```python
import matplotlib.pyplot as plt
import sys
import pandas as pd
import numpy as np
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
import glob
import seaborn as sns
import mdtraj as md
import nglview as ngl
```

```python
def plot_potential(file):
    fig, ax = plt.subplots(1,1,figsize = (2.5,2))
    min_ = pd.read_csv(file, sep='\s+',skiprows=24,names=['Time (ps)','Potential (kJ/mol)'])
    plt.plot(min_['Time (ps)'],min_['Potential (kJ/mol)'])
    ax.set_ylabel('Potential (kJ/mol)')
    ax.set_xlabel('Time (ps)')
def plot_temp(file):
    fig, ax = plt.subplots(1,1,figsize = (2.5,2))
    min_ = pd.read_csv(file, sep='\s+',skiprows=24,names=['Time (ps)','Temperature (K)'])
    plt.plot(min_['Time (ps)'],min_['Temperature (K)'])
    ax.set_ylabel('Temperature (K)')
    ax.set_xlabel('Time (ps)')
def plot_dens(file):
    fig, ax = plt.subplots(1,1,figsize = (2.5,2))
    min_ = pd.read_csv(file, sep='\s+',skiprows=24,names=['Time (ps)','Density (kg/m^3)'])
    plt.plot(min_['Time (ps)'],min_['Density (kg/m^3)'])
    ax.set_ylabel('Density (kg/m^3)')
    ax.set_xlabel('Time (ps)')
def plot_pressure(file):
    fig, ax = plt.subplots(1,1,figsize = (2.5,2))
    min_ = pd.read_csv(file, sep='\s+',skiprows=24,names=['Time (ps)','Pressure (bar)'])
    plt.plot(min_['Time (ps)'],min_['Pressure (bar)'])
    ax.set_ylabel('Pressure (bar)')
    ax.set_xlabel('Time (ps)')
```
